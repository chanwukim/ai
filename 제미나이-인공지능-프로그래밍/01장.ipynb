```python
import typing_extensions as typing
from google.colab import userdata
import google.generativeai as genai

class Recipe(typing.TypedDict):
  recipe_name: str

genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

model = genai.GenerativeModel(
    model_name="models/gemini-2.5-flash"
)

result = model.generate_content(
    "ìœ ë¬­í•œ ì¿ í‚¤ ë ˆì‹œí”¼ë¥¼ í•œêµ­ì–´ë¡œ ë‹¤ì„¯ ê°œ ë¦¬ìŠ¤íŠ¸ì—… í•´ì£¼ì„¸ìš”.",
    generation_config=genai.GenerationConfig(
        response_mime_type="application/json",
        response_schema=list[Recipe]
    ),
    request_options={"timeout": 600}
)

print(type(result.text))
print(result.text)
```

# Function calling

```python
def add(a:float, b:float):
  """returns a + b."""
  return a+b

def subtract(a:float, b:float):
  """returns a - b."""
  return a-b

def multiply(a:float, b:float):
  """returns a * b."""
  return a*b

def divide(a:float, b:float):
  """returns a / b."""
  return a/b

model = genai.GenerativeModel(
    model_name="models/gemini-2.5-flash",
    tools=[add, subtract, multiply, divide]
)

chat = model.start_chat(
    enable_automatic_function_calling=True
)

response = chat.send_message(
    "ì €ëŠ” 57ë§ˆë¦¬ì˜ ê³ ì–‘ì´ë¥¼ í‚¤ìš°ê³  ìˆê³ , ê° 44ê°œì˜ ì†ì‹¸ê°œë¥¼ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. ì†ì‹¸ê°œëŠ” ì´ ëª‡ê°œì¼ê¹Œìš”?"
)
print(response.text)

for content in chat.history:
  print(content.role, "->", [type(part).to_dict(part) for part in content.parts])
  print("--")
```

## ë„êµ¬ ì„¤ì •í•˜ê¸°
ì‚¬ìš©ì ì…ë ¥ì— ë”°ë¥¸ ë©”ì‹œì§€ ì‘ë‹µê³¼ í•¨ìˆ˜ í˜¸ì¶œ ì„ íƒì„ ì œì–´í•˜ëŠ” ì„¤ì •
- NONE: ë©”ì‹œì§€ ì‘ë‹µ ê°•ì œ
- AUTO: ë©”ì‹œì§€ ì‘ë‹µ ë° í•¨ìˆ˜ í˜¸ì¶œì„ ìë™ìœ¼ë¡œ ì„ íƒ
- ANY: í•¨ìˆ˜ í˜¸ì¶œ ê°•ì œ

```python
from google.generativeai.types import content_types
from collections.abc import Iterable

def tool_config_from_mode(mode:str, fns: Iterable[str] = ()):
  """í•¨ìˆ˜ í˜¸ì¶œ ëª¨ë“œì™€ í—ˆìš©í•˜ëŠ” í•¨ìˆ˜ ì´ë¦„ìœ¼ë¡œ ë„êµ¬ ì„¤ì •"""

  return content_types.to_tool_config(
    {"function_calling_config": {"mode": mode, "allowed_function_names": fns}}
)

chat.history.clear()

response = chat.send_message(
    "ì €ëŠ” 57ë§ˆë¦¬ì˜ ê³ ì–‘ì´ë¥¼ í‚¤ìš°ê³  ìˆê³ , ê° 44ê°œì˜ ì†ì‹¸ê°œë¥¼ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. ì†ì‹¸ê°œëŠ” ì´ ëª‡ê°œì¼ê¹Œìš”?",
    tool_config=tool_config_from_mode("auto")
)
print(response.text)


for content in chat.history:
  print(content.role, "->", [type(part).to_dict(part) for part in content.parts])
  print("--")
```
